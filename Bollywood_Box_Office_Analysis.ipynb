# Bollywood Box Office Success Analysis: A Data-Driven Study of Revenue, Budget, Genre, and Star Power
## Objective: Bollywood Box Office Success Analysis: A Data-Driven Study of Revenue, Budget, Genre, and Star Power

```python
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('Data for repository.csv')

df.head()
```

```python
print(df.columns.tolist())
```

```python
top10 = df.nlargest(10, 'Revenue(INR)')[['Movie_Name', 'Revenue(INR)']]
plt.figure(figsize=(10,6))
sns.barplot(data=top10, x='Revenue(INR)', y='Movie_Name')
plt.title('Top 10 Highest Revenue Bollywood Movies')
plt.show()
```

### Tiger Zinda Hai generated the highest revenue in the dataset, followed by Bahubali 2: The Conclusion. The graph shows that a few blockbuster movies dominate box office collections.

```python
sns.scatterplot(data=df, x='Budget(INR)', y= 'Revenue(INR)' )
plt.title('Budget vs Revenue')
plt.show()
```

### Movies with larger budgets generally generate higher revenue. However, budget alone does not guarantee success, as several films show significantly different revenues despite having similar budgets.

```python
genre_rev = df.groupby('Genre')['Revenue(INR)'].mean().sort_values()

plt.figure(figsize=(10,6))
genre_rev.plot(kind='barh')
plt.title('Average Revenue by Genre')
plt.show()
```

### Masala films achieved the highest average revenue among all genres, whereas Adult films generated the lowest. This suggests that mainstream commercial movies tend to perform better at the box office than niche genres

```python
sns.boxplot(data=df,
            x='Whether_Franchise',
            y='Revenue(INR)')
plt.show()
```

### Franchise movies generally earned higher revenue than non-franchise movies. This suggests that established franchises tend to attract larger audiences.

```python
sns.boxplot(data=df,
            x='Whether_Remake',
            y='Revenue(INR)')
plt.show()
```

### Non-remake films have more concentrated revenue distributions, whereas some remake films generated exceptionally high revenues. This suggests that remakes have the potential for blockbuster success but do not consistently outperform original films.

```python
sns.regplot(data=df,
            x='Number_of_Screens',
            y='Revenue(INR)')
plt.show()
```

### Revenue tends to increase as the number of screens increases. This indicates that wider theatrical releases are associated with better box office performance.

```python
top_stars = df.groupby('Lead_Star')['Revenue(INR)'].mean().sort_values(ascending=False).head(10)

top_stars.plot(kind='bar')
plt.title('Top Lead Stars by Average Revenue')
plt.show()
```

### Prabhas recorded the highest average revenue among all lead stars. This highlights his strong commercial performance and popularity in the dataset.

```python
num_cols = ['Budget(INR)',
            'Revenue(INR)',
            'Number_of_Screens']

sns.heatmap(df[num_cols].corr(),
            annot=True,
            cmap='coolwarm')
plt.show()
```

### Higher movie budgets are strongly linked to more screens and higher revenue-bigger investments,bigger reach and returns

### Summary : The analysis shows that movie revenue is influenced by multiple factors, including budget, genre, franchise status, screen count, and lead actors. Commercial genres and franchise films generally performed better, while wider theatrical releases were associated with higher revenues. These findings highlight the value of data analysis in understanding box office trends.
